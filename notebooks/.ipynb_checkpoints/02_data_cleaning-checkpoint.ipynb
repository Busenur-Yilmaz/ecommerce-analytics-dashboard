```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(" Kütüphaneler yüklendi!")
print("\n" + "=" * 80)
print(" VERİ TEMİZLEME SÜRECİ ")
print("=" * 80)
```

```python
RAW_DATA_PATH = '../data/raw/'

print("\n Veri setleri yükleniyor...")

orders = pd.read_csv(f'{RAW_DATA_PATH}olist_orders_dataset.csv')
customers = pd.read_csv(f'{RAW_DATA_PATH}olist_customers_dataset.csv')
order_items = pd.read_csv(f'{RAW_DATA_PATH}olist_order_items_dataset.csv')
products = pd.read_csv(f'{RAW_DATA_PATH}olist_products_dataset.csv')
sellers = pd.read_csv(f'{RAW_DATA_PATH}olist_sellers_dataset.csv')
payments = pd.read_csv(f'{RAW_DATA_PATH}olist_order_payments_dataset.csv')
reviews = pd.read_csv(f'{RAW_DATA_PATH}olist_order_reviews_dataset.csv')
geolocation = pd.read_csv(f'{RAW_DATA_PATH}olist_geolocation_dataset.csv')

print(" Tüm veri setleri yüklendi!")
print(f"\n Yüklenen veri setleri:")
print(f"   - Orders: {orders.shape}")
print(f"   - Customers: {customers.shape}")
print(f"   - Order Items: {order_items.shape}")
print(f"   - Products: {products.shape}")
print(f"   - Sellers: {sellers.shape}")
print(f"   - Payments: {payments.shape}")
print(f"   - Reviews: {reviews.shape}")
print(f"   - Geolocation: {geolocation.shape}")
```

```python
print("\n" + "=" * 80)
print("1️-ORDERS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(orders):,}")
print(f"Eksik değerler:\n{orders.isnull().sum()}")

# Tarih kolonlarını datetime'a çeviriyoruz
date_columns = ['order_purchase_timestamp', 'order_approved_at', 
                'order_delivered_carrier_date', 'order_delivered_customer_date',
                'order_estimated_delivery_date']

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

# Sadece tamamlanmış siparişleri alıyoruz (delivered)
orders_clean = orders[orders['order_status'] == 'delivered'].copy()

# Yıl, ay, çeyrek bilgilerini ekle
orders_clean['order_year'] = orders_clean['order_purchase_timestamp'].dt.year
orders_clean['order_month'] = orders_clean['order_purchase_timestamp'].dt.month
orders_clean['order_quarter'] = orders_clean['order_purchase_timestamp'].dt.quarter
orders_clean['order_year_month'] = orders_clean['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Teslimat süresi hesaplıyoruz (gün olarak)
orders_clean['delivery_time_days'] = (
    orders_clean['order_delivered_customer_date'] - 
    orders_clean['order_purchase_timestamp']
).dt.days

# Tahmini teslimat farkı (gerçek - tahmini)
orders_clean['delivery_vs_estimated'] = (
    orders_clean['order_delivered_customer_date'] - 
    orders_clean['order_estimated_delivery_date']
).dt.days

# Negatif teslimat sürelerini temizliyoruz (veri hatası)
orders_clean = orders_clean[orders_clean['delivery_time_days'] >= 0]

print("\n Temizleme sonrası durum:")
print(f"Toplam satır: {len(orders_clean):,}")
print(f"Silinen satır: {len(orders) - len(orders_clean):,}")
print(f"Yeni kolonlar: order_year, order_month, order_quarter, order_year_month, delivery_time_days, delivery_vs_estimated")
print(f"\nOrtalama teslimat süresi: {orders_clean['delivery_time_days'].mean():.1f} gün")
print(f"Medyan teslimat süresi: {orders_clean['delivery_time_days'].median():.1f} gün")
```

```python
print("\n" + "=" * 80)
print("2️-CUSTOMERS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(customers):,}")
print(f"Eksik değerler:\n{customers.isnull().sum()}")

# Duplikaları kontrol et
duplicates = customers.duplicated(subset=['customer_id']).sum()
print(f"\nDuplikat customer_id sayısı: {duplicates}")

customers_clean = customers.copy()

# Şehir ve eyalet isimlerini standartlaştır (büyük harf)
customers_clean['customer_city'] = customers_clean['customer_city'].str.upper()
customers_clean['customer_state'] = customers_clean['customer_state'].str.upper()

# Posta kodunu integer'a çevir
customers_clean['customer_zip_code_prefix'] = customers_clean['customer_zip_code_prefix'].astype(int)

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(customers_clean):,}")
print(f"Benzersiz müşteri: {customers_clean['customer_unique_id'].nunique():,}")
```

```python
print("\n" + "=" * 80)
print("3️-ORDER ITEMS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(order_items):,}")
print(f"Eksik değerler:\n{order_items.isnull().sum()}")

order_items_clean = order_items.copy()

# Negatif veya sıfır fiyatları temizle
print(f"\nNegatif/sıfır fiyat sayısı: {(order_items_clean['price'] <= 0).sum()}")
order_items_clean = order_items_clean[order_items_clean['price'] > 0]

# Toplam sipariş değeri hesapla (ürün + kargo)
order_items_clean['total_price'] = order_items_clean['price'] + order_items_clean['freight_value']

# Shipping tarihini datetime'a çevir
order_items_clean['shipping_limit_date'] = pd.to_datetime(
    order_items_clean['shipping_limit_date'], errors='coerce'
)

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(order_items_clean):,}")
print(f"Silinen satır: {len(order_items) - len(order_items_clean):,}")
print(f"Yeni kolon: total_price")
print(f"\nOrtalama ürün fiyatı: ${order_items_clean['price'].mean():.2f}")
print(f"Ortalama kargo: ${order_items_clean['freight_value'].mean():.2f}")
print(f"Ortalama toplam: ${order_items_clean['total_price'].mean():.2f}")
```

```python
print("\n" + "=" * 80)
print("4️-PRODUCTS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(products):,}")
print(f"Eksik değerler:\n{products.isnull().sum()}")

products_clean = products.copy()

# Eksik kategori isimlerini "unknown" ile doldur
products_clean['product_category_name'].fillna('unknown', inplace=True)

# Ürün boyutlarında eksik değerleri medyan ile doldur
dimension_cols = ['product_weight_g', 'product_length_cm', 
                  'product_height_cm', 'product_width_cm']

for col in dimension_cols:
    median_val = products_clean[col].median()
    products_clean[col].fillna(median_val, inplace=True)

# Ürün hacmi hesapla (cm³)
products_clean['product_volume_cm3'] = (
    products_clean['product_length_cm'] * 
    products_clean['product_height_cm'] * 
    products_clean['product_width_cm']
)

# Ağırlığı kg'ye çevir
products_clean['product_weight_kg'] = products_clean['product_weight_g'] / 1000

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(products_clean):,}")
print(f"Yeni kolonlar: product_volume_cm3, product_weight_kg")
print(f"\nKategori sayısı: {products_clean['product_category_name'].nunique()}")
```

```python
print("\n" + "=" * 80)
print("5️-SELLERS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(sellers):,}")
print(f"Eksik değerler:\n{sellers.isnull().sum()}")

sellers_clean = sellers.copy()

# Şehir ve eyalet isimlerini standartlaştır
sellers_clean['seller_city'] = sellers_clean['seller_city'].str.upper()
sellers_clean['seller_state'] = sellers_clean['seller_state'].str.upper()

# Posta kodunu integer'a çevir
sellers_clean['seller_zip_code_prefix'] = sellers_clean['seller_zip_code_prefix'].astype(int)

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(sellers_clean):,}")
print(f"Benzersiz satıcı: {sellers_clean['seller_id'].nunique():,}")
```

```python
print("\n" + "=" * 80)
print("6️-PAYMENTS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(payments):,}")
print(f"Eksik değerler:\n{payments.isnull().sum()}")

payments_clean = payments.copy()

# Negatif veya sıfır ödemeleri temizle
print(f"\nNegatif/sıfır ödeme sayısı: {(payments_clean['payment_value'] <= 0).sum()}")
payments_clean = payments_clean[payments_clean['payment_value'] > 0]

# Ödeme tiplerini standartlaştır
payments_clean['payment_type'] = payments_clean['payment_type'].str.lower()

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(payments_clean):,}")
print(f"Silinen satır: {len(payments) - len(payments_clean):,}")
print(f"\nÖdeme tipleri:")
print(payments_clean['payment_type'].value_counts())
```

```python
print("\n" + "=" * 80)
print("7️-REVIEWS VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(reviews):,}")
print(f"Eksik değerler:\n{reviews.isnull().sum()}")

reviews_clean = reviews.copy()

# Tarih kolonlarını datetime'a çevir
reviews_clean['review_creation_date'] = pd.to_datetime(
    reviews_clean['review_creation_date'], errors='coerce'
)
reviews_clean['review_answer_timestamp'] = pd.to_datetime(
    reviews_clean['review_answer_timestamp'], errors='coerce'
)

# Score'u 1-5 aralığına sınırla
reviews_clean = reviews_clean[reviews_clean['review_score'].between(1, 5)]

# Yorum ve başlık için eksik değerleri doldur
reviews_clean['review_comment_title'].fillna('', inplace=True)
reviews_clean['review_comment_message'].fillna('', inplace=True)

# Yorum uzunluğu hesapla
reviews_clean['comment_length'] = reviews_clean['review_comment_message'].str.len()

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(reviews_clean):,}")
print(f"Ortalama puan: {reviews_clean['review_score'].mean():.2f}/5.00")
print(f"Yorumlu review oranı: {(reviews_clean['comment_length'] > 0).mean()*100:.1f}%")
```

```python
print("\n" + "=" * 80)
print("8️-GEOLOCATION VERİ SETİNİ TEMİZLE")
print("=" * 80)

print("\n Önceki durum:")
print(f"Toplam satır: {len(geolocation):,}")
print(f"Eksik değerler:\n{geolocation.isnull().sum()}")

# Geolocation çok büyük, zip code başına ortalama koordinatlar alalım
geolocation_clean = geolocation.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean',
    'geolocation_city': 'first',
    'geolocation_state': 'first'
}).reset_index()

# Şehir ve eyalet standartlaştır
geolocation_clean['geolocation_city'] = geolocation_clean['geolocation_city'].str.upper()
geolocation_clean['geolocation_state'] = geolocation_clean['geolocation_state'].str.upper()

print("\n Temizleme tamamlandı!")
print(f"Toplam satır: {len(geolocation):,} → {len(geolocation_clean):,}")
print(f"Benzersiz zip code: {geolocation_clean['geolocation_zip_code_prefix'].nunique():,}")
```

```python
print("\n" + "=" * 80)
print(" TEMİZLENMİŞ VERİLER KAYDEDİLİYOR")
print("=" * 80)

PROCESSED_PATH = '../data/processed/'

# Her veri setini ayrı ayrı kaydet
orders_clean.to_csv(f'{PROCESSED_PATH}orders_clean.csv', index=False)
customers_clean.to_csv(f'{PROCESSED_PATH}customers_clean.csv', index=False)
order_items_clean.to_csv(f'{PROCESSED_PATH}order_items_clean.csv', index=False)
products_clean.to_csv(f'{PROCESSED_PATH}products_clean.csv', index=False)
sellers_clean.to_csv(f'{PROCESSED_PATH}sellers_clean.csv', index=False)
payments_clean.to_csv(f'{PROCESSED_PATH}payments_clean.csv', index=False)
reviews_clean.to_csv(f'{PROCESSED_PATH}reviews_clean.csv', index=False)
geolocation_clean.to_csv(f'{PROCESSED_PATH}geolocation_clean.csv', index=False)

print("\n Tüm temizlenmiş veri setleri kaydedildi!")
print(f"\n Kayıt lokasyonu: {PROCESSED_PATH}")
print("""
Kaydedilen dosyalar:
  ✓ orders_clean.csv
  ✓ customers_clean.csv
  ✓ order_items_clean.csv
  ✓ products_clean.csv
  ✓ sellers_clean.csv
  ✓ payments_clean.csv
  ✓ reviews_clean.csv
  ✓ geolocation_clean.csv
""")
```

```python
print("\n" + "=" * 80)
print(" TEMİZLEME ÖZETİ")
print("=" * 80)

summary_data = {
    'Dataset': ['Orders', 'Customers', 'Order Items', 'Products', 
                'Sellers', 'Payments', 'Reviews', 'Geolocation'],
    'Önceki': [len(orders), len(customers), len(order_items), len(products),
               len(sellers), len(payments), len(reviews), len(geolocation)],
    'Sonraki': [len(orders_clean), len(customers_clean), len(order_items_clean),
                len(products_clean), len(sellers_clean), len(payments_clean),
                len(reviews_clean), len(geolocation_clean)]
}

summary_df = pd.DataFrame(summary_data)
summary_df['Silinen'] = summary_df['Önceki'] - summary_df['Sonraki']
summary_df['Kayıp %'] = (summary_df['Silinen'] / summary_df['Önceki'] * 100).round(2)

print("\n")
print(summary_df.to_string(index=False))

print("\n" + "=" * 80)
print(" VERİ TEMİZLEME TAMAMLANDI!")
print("=" * 80)
print("""
 YAPILAN İŞLEMLER:
  ✓ Tarih formatları düzeltildi
  ✓ Eksik değerler dolduruldu
  ✓ Negatif/hatalı değerler temizlendi
  ✓ Yeni feature'lar oluşturuldu
  ✓ Standartlaştırmalar yapıldı
  ✓ Temizlenmiş veriler kaydedildi""")
```
