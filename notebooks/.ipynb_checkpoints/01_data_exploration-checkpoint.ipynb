```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(" Kütüphaneler başarıyla yüklendi!")
print(f" Pandas version: {pd.__version__}")
print(f" Numpy version: {np.__version__}")
```

```python
RAW_DATA_PATH = '../data/raw/'

print("=" * 80)
print("BRAZILIAN E-COMMERCE VERİ SETİ KEŞFİ")
print("=" * 80)
```

```python
print("\n 1️-ORDERS DATASET")
print("-" * 80)

orders = pd.read_csv(f'{RAW_DATA_PATH}olist_orders_dataset.csv')

print(f"Boyut: {orders.shape}")
print(f"\nKolonlar: {orders.columns.tolist()}")
print(f"\nİlk 5 satır:")
display(orders.head())
print(f"\nVeri tipleri:")
print(orders.dtypes)
print(f"\nEksik değerler:")
print(orders.isnull().sum())
print(f"\nSipariş durumları:")
print(orders['order_status'].value_counts())
```

```python
print("\n 2-CUSTOMERS DATASET")
print("-" * 80)

customers = pd.read_csv(f'{RAW_DATA_PATH}olist_customers_dataset.csv')

print(f"Boyut: {customers.shape}")
print(f"\nKolonlar: {customers.columns.tolist()}")
print(f"\nBenzersiz müşteri sayısı: {customers['customer_unique_id'].nunique():,}")
print(f"Benzersiz şehir sayısı: {customers['customer_city'].nunique():,}")
print(f"Benzersiz eyalet sayısı: {customers['customer_state'].nunique():,}")

print(f"\nİlk 5 satır:")
display(customers.head())

print(f"\nEn çok müşteri olan 10 şehir:")
print(customers['customer_city'].value_counts().head(10))
```

```python
print("\n 3-ORDER ITEMS DATASET")
print("-" * 80)

order_items = pd.read_csv(f'{RAW_DATA_PATH}olist_order_items_dataset.csv')

print(f"Boyut: {order_items.shape}")
print(f"\nKolonlar: {order_items.columns.tolist()}")
print(f"\nToplam satış tutarı: ${order_items['price'].sum():,.2f}")
print(f"Toplam kargo tutarı: ${order_items['freight_value'].sum():,.2f}")
print(f"Ortalama ürün fiyatı: ${order_items['price'].mean():.2f}")

print(f"\nİlk 5 satır:")
display(order_items.head())

print(f"\nFiyat istatistikleri:")
print(order_items['price'].describe())
```

```python
print("\n 4-PRODUCTS DATASET")
print("-" * 80)

products = pd.read_csv(f'{RAW_DATA_PATH}olist_products_dataset.csv')

print(f"Boyut: {products.shape}")
print(f"\nKolonlar: {products.columns.tolist()}")
print(f"\nBenzersiz ürün sayısı: {products['product_id'].nunique():,}")
print(f"Benzersiz kategori sayısı: {products['product_category_name'].nunique():,}")

print(f"\nİlk 5 satır:")
display(products.head())

print(f"\nEn popüler 10 kategori:")
print(products['product_category_name'].value_counts().head(10))

print(f"\nEksik değerler:")
print(products.isnull().sum())
```

```python
print("\n 5-SELLERS DATASET")
print("-" * 80)

sellers = pd.read_csv(f'{RAW_DATA_PATH}olist_sellers_dataset.csv')

print(f"Boyut: {sellers.shape}")
print(f"\nKolonlar: {sellers.columns.tolist()}")
print(f"\nBenzersiz satıcı sayısı: {sellers['seller_id'].nunique():,}")
print(f"Satıcıların bulunduğu şehir sayısı: {sellers['seller_city'].nunique():,}")
print(f"Satıcıların bulunduğu eyalet sayısı: {sellers['seller_state'].nunique():,}")

print(f"\nİlk 5 satır:")
display(sellers.head())

print(f"\nEn çok satıcı olan 10 şehir:")
print(sellers['seller_city'].value_counts().head(10))
```

```python
print("\n 6-PAYMENTS DATASET")
print("-" * 80)

payments = pd.read_csv(f'{RAW_DATA_PATH}olist_order_payments_dataset.csv')

print(f"Boyut: {payments.shape}")
print(f"\nKolonlar: {payments.columns.tolist()}")
print(f"\nToplam ödeme tutarı: ${payments['payment_value'].sum():,.2f}")

print(f"\nİlk 5 satır:")
display(payments.head())

print(f"\nÖdeme tipleri ve sayıları:")
print(payments['payment_type'].value_counts())

print(f"\nÖdeme tiplerine göre toplam tutar:")
payment_by_type = payments.groupby('payment_type')['payment_value'].sum().sort_values(ascending=False)
print(payment_by_type)

# Ödeme tiplerini görselleştir
plt.figure(figsize=(10, 6))
payment_by_type.plot(kind='bar', color='skyblue')
plt.title('Ödeme Tiplerine Göre Toplam Tutar', fontsize=14, fontweight='bold')
plt.xlabel('Ödeme Tipi')
plt.ylabel('Toplam Tutar ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\n 7-REVIEWS DATASET")
print("-" * 80)

reviews = pd.read_csv(f'{RAW_DATA_PATH}olist_order_reviews_dataset.csv')

print(f"Boyut: {reviews.shape}")
print(f"\nKolonlar: {reviews.columns.tolist()}")
print(f"\nOrtalama puan: {reviews['review_score'].mean():.2f}/5.00")

print(f"\nİlk 5 satır:")
display(reviews.head())

print(f"\nPuan dağılımı:")
score_dist = reviews['review_score'].value_counts().sort_index()
print(score_dist)

# Review skorlarını görselleştir
plt.figure(figsize=(10, 6))
score_dist.plot(kind='bar', color='coral')
plt.title('Müşteri Puanı Dağılımı', fontsize=14, fontweight='bold')
plt.xlabel('Puan')
plt.ylabel('Sayı')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
print("\n 8-GEOLOCATION DATASET")
print("-" * 80)

geolocation = pd.read_csv(f'{RAW_DATA_PATH}olist_geolocation_dataset.csv')

print(f"Boyut: {geolocation.shape}")
print(f"\nKolonlar: {geolocation.columns.tolist()}")
print(f"\nBenzersiz posta kodu sayısı: {geolocation['geolocation_zip_code_prefix'].nunique():,}")

print(f"\nİlk 5 satır:")
display(geolocation.head())
```

```python
print("\n" + "=" * 80)
print(" TARİH ARALIĞI ANALİZİ")
print("=" * 80)

# Tarih kolonlarını datetime'a çevir
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

print(f"\nİlk sipariş: {orders['order_purchase_timestamp'].min()}")
print(f"Son sipariş: {orders['order_purchase_timestamp'].max()}")
print(f"Toplam süre: {(orders['order_purchase_timestamp'].max() - orders['order_purchase_timestamp'].min()).days} gün")

# Aylık sipariş trendi
orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
monthly_orders = orders.groupby('year_month').size()

plt.figure(figsize=(14, 6))
monthly_orders.plot(kind='line', marker='o', color='green')
plt.title('Aylık Sipariş Trendi', fontsize=14, fontweight='bold')
plt.xlabel('Ay')
plt.ylabel('Sipariş Sayısı')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 80)
print("GENEL ÖZET İSTATİSTİKLER")
print("=" * 80)

# Toplam gelir hesapla
total_revenue = payments['payment_value'].sum()
avg_order_value = payments.groupby('order_id')['payment_value'].sum().mean()

print(f"""
 Toplam Sipariş: {len(orders):,}
 Tamamlanan Sipariş: {(orders['order_status'] == 'delivered').sum():,}
 Toplam Müşteri: {customers['customer_unique_id'].nunique():,}
 Toplam Ürün: {products['product_id'].nunique():,}
 Toplam Satıcı: {sellers['seller_id'].nunique():,}
 Toplam Kategori: {products['product_category_name'].dropna().nunique():,}
 Toplam Gelir: ${total_revenue:,.2f}
 Ortalama Sipariş Değeri: ${avg_order_value:.2f}
 Ortalama Müşteri Puanı: {reviews['review_score'].mean():.2f}/5.00
 Veri Zaman Aralığı: {orders['order_purchase_timestamp'].min().date()} - {orders['order_purchase_timestamp'].max().date()}
""")
```

```python
print("\n" + "=" * 80)
print(" VERİ SETLERİ ARASINDAKİ İLİŞKİLER")
print("=" * 80)

print("""
TEMEL İLİŞKİLER:
┌─────────────────────────────────────────────────────────┐
│ orders (order_id) ←→ order_items (order_id)             │
│ orders (customer_id) ←→ customers (customer_id)         │
│ order_items (product_id) ←→ products (product_id)       │
│ order_items (seller_id) ←→ sellers (seller_id)          │
│ orders (order_id) ←→ payments (order_id)                │
│ orders (order_id) ←→ reviews (order_id)                 │
│ customers (zip_code) ←→ geolocation (zip_code)          │
└─────────────────────────────────────────────────────────┘
""")

# Join kalitesini kontrol et
print("\nJOIN KALİTE KONTROLLERI:")
print(f"✓ Orders'da order_items ile eşleşen: {orders['order_id'].isin(order_items['order_id']).sum():,}/{len(orders):,}")
print(f"✓ Orders'da customers ile eşleşen: {orders['customer_id'].isin(customers['customer_id']).sum():,}/{len(orders):,}")
print(f"✓ Order_items'da products ile eşleşen: {order_items['product_id'].isin(products['product_id']).sum():,}/{len(order_items):,}")
print(f"✓ Order_items'da sellers ile eşleşen: {order_items['seller_id'].isin(sellers['seller_id']).sum():,}/{len(order_items):,}")
```
